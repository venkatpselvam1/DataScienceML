#### Columns
- **instant** - It is runnings number from 1 to 730 (2 year = 365 days * 2)  - It is not going to help us in the prediction. **So we can remove it. Action1**
- **season** - It indicates the season of the month. (1:spring => 180, 2:summer => 184, 3:fall => 188, 4:winter => 178). **So we need map the column. Action2**
- **dteday** - It indicates the date of the row. It is unique values b/w the date range[ from 2018-01-01 to 2019-12-31] . **We can convert it to datetime object and further number from 1 to 720. But the columns month, season, weekday, holiday, everything is already extracted we don't need this column.   Action1: So we can remove it.**
- **yr** - It is categorical variable with value 0 indicates 2018 and 1 indicates 2019. We have 365 times 1 value and 365 times 0 value.
- **mnth** - It is again categorical value with the value 1 to 12 indicates each month
- **holiday** - It indicates if it is holiday. There are 21 holiday(with value 1) and 709 working days(with value 0)
- **weekday** - It indicates the day of the week 0 to 6. [0 - Sunday, 1 - Monday, 2 - Tuesday, 3 - Wednesday, 4 - Thursday, 5 - Friday, 6 - Saturday]
- **workingday** - It incluses the (21 holiday + 105 Saturday + 104 Sundays ). We have total 504 - Working Day and 226 leaves
- **temp, atemp, hum, windspeed** - These are the continuous numerical values
- **weathersit** - Have value( 1:Clear => 463 , 2:Mist => 246 , 3:Snow => 21, 4: HeavyRain => 0 ) **So we need map the column. Action2**
- **casual + registered = cnt (label - target value)** - We want to predict the cnt. The CNT is the count of two continuouse variables casual + registered. **Action 4: We can create a new feature saying casual_registered_ratio which indicates the => casual/registered**

```python
#Action 1: Remove instant column
df.drop('instant', axis=1, inplace=True)
df.drop('dteday', axis=1, inplace=True)

#Action 2: Map the columns weathersit and season
df['season'] = df['season'].map({1:'spring',2:'summer',3:'fall',4:'winter'})
df['weathersit'] = df['weathersit'].map({1:'clear_cloud',2:'mist',3:'snow',4:'heavy_rain'})

#Actino 4: adding new casual_registered_ratio and removing casual and registered columns
df['casual_registered_ratio'] = df['casual']/df['registered']
df.drop('casual', axis=1, inplace=True)
df.drop('registered', axis=1, inplace=True)
```

## Scaling
- In this problem we will use MinMaxScaler to normalize the numerical columns
- But before normalizing the data, we need to split the test data into test and train
- Because we want to apply the same scaler(which we created for train data set) to the test data set

```python
import statsmodels
import statsmodels.api as sm
import sklearn
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
```

```python
df_train, df_test = train_test_split(df, test_size=0.3, random_state=100)
print("checkig the size after splitting the train and test")
print("Original shape = ", df.shape, " Train shape = ", df_train.shape, " Test shape = ", df_test.shape)
```

```python
num_cols = ['temp', 'hum', 'windspeed', 'cnt', 'casual_registered_ratio']
df_train[num_cols].describe()
```

```python
scaler = MinMaxScaler()
df_train[num_cols] = scaler.fit_transform(df_train[num_cols])
```

```python
df_train[num_cols].describe()
```

```python
df_test[num_cols].describe()
```

```python
df_test[num_cols] = scaler.transform(df_test[num_cols])
```

```python
df_test[num_cols].describe()
```

- Notice that after the normalization the train data set's range is between 0 and 1
- For the test dataset, we just used the scaler from the train data, we notice that the min and max is varies

## Split the df to X and y for the model training and evaluation

```python
y_train = df_train.pop('cnt')
X_train = df_train
y_test = df_test.pop('cnt')
X_test = df_test
print("X_train = ", X_train.shape, ", y_train = ", y_train.shape, "X_test = ", X_test.shape, ", y_test=", y_test.shape)
```

# Create Linear Regression model using Sklearn's RFE (Recursive feature elemination)

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(lm)
rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
cols = X_train.columns[ rfe.support_]
cols
```

```python
rejected_cols = X_train.columns[ ~rfe.support_]
rejected_cols
```

```python
X = X_train.drop(rejected_cols, axis=True)
X_train_sm = sm.add_constant(X)
lm = sm.OLS(y_train, X_train_sm)
lr_model = lm.fit()
lr_model.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

# Observation from RFE
- We used sklearn's RFE to do the feature selection
- RFE selected ['yr', 'holiday', 'temp', 'hum', 'windspeed', 'spring', 'Snow'] columns
- RFE rejected ['mnth', 'weekday', 'workingday', 'casual_registered_ratio', 'summer', 'winter', 'Mist'] columns
- Using these selected columns we built the linear regression model using stats model
- From statsmodel's summary -> R^2 and adjusted R^2 are areound 0.8 which is good.
- From statsmodel's summary -> The p value for all the selected columns are within the 0.05 which indicates the variables are significant
- From statsmodel's VIF calculation -> We see the hum and temperature have higher VIFs 9.76 and 8.41 respectively
- We need to drop the column one by one. First we drop teh hum column. If the VIF of temp is still more, we need to drop the temp also

```python
total_rejected_cols = list(rejected_cols) + ['hum']
```

```python
X = X_train.drop(total_rejected_cols, axis=True)
X_train_sm = sm.add_constant(X)
lm = sm.OLS(y_train, X_train_sm)
lr_model = lm.fit()
lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

# Observation from RFE
- Notice that after removing the column hum the VIF of temp also came below 5. 
- All the columns' VIF is less than 5
- From statsmodel's summary -> R^2 = 0.79 and adjusted R^2 =  0.78 which are good.
- From statsmodel's summary -> The p value for all the selected columns are within the 0.05 which indicates the variables are significant

# Model Evaluation

## Model evaluation using train data set - Residual Analysis

```python
y_train_pred = lr_model.predict(X_train_sm)
res = y_train-y_train_pred
fig = plt.figure()
sns.displot(res)
fig.suptitle('Error Terms') 
plt.xlabel('Errors', fontsize = 18) 
```

## Model evaluation using test data set - Residual Analysis

```python
X = X_test.drop(total_rejected_cols, axis=True)
X_test_sm = sm.add_constant(X)
y_test_pred = lr_model.predict(X_test_sm)
res = y_test-y_test_pred
fig = plt.figure()
sns.displot(res)
fig.suptitle('Error Terms') 
plt.xlabel('Errors', fontsize = 18)
```

### for the train and  test data set also notice that error terms are normally distributted
- From the above two displot notice that error terms are normally distributed
- So it passes the assumption of Linear Regression

## Model Evaluation

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
print("train r2_score = ", r2_score(y_train, y_train_pred))
print("test r2_score = ", r2_score(y_test, y_test_pred))
print("train mean_squared_error = ", mean_squared_error(y_train, y_train_pred))
print("test mean_squared_error = ", mean_squared_error(y_test, y_test_pred))
```
