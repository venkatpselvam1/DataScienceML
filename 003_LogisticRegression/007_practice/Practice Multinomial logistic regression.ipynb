```python
Y = df.pop('quality')
X = df
```

```python
import statsmodels 
import statsmodels.api as sm
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.3,random_state = 10)
LR = LogisticRegression()
OneVsRes = OneVsRestClassifier(LR)
# Fitting the model with training data
OneVsRes.fit(X_train, y_train)
```

```python
prediction_one_vs_rest_test = OneVsRes.predict(X_test)
prediction_one_vs_rest_train = OneVsRes.predict(X_train)
```

```python
print("Test accuracy " ,  accuracy_score(y_test, prediction_one_vs_rest_test))
```

```python
print(f"Classification Report: \n\n{classification_report(y_train, prediction_one_vs_rest_train)}")
```
