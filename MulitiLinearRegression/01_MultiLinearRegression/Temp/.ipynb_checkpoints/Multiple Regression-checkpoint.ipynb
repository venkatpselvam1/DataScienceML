```python
df['guestroom'] = df['guestroom'].map({'yes':1, 'no':0})
df['mainroad'] = df['mainroad'].map({'yes':1, 'no':0})
df['basement'] = df['basement'].map({'yes':1, 'no':0})
df['hotwaterheating'] = df['hotwaterheating'].map({'yes':1, 'no':0})
df['airconditioning'] = df['airconditioning'].map({'yes':1, 'no':0})
df['prefarea'] = df['prefarea'].map({'yes':1, 'no':0})
```

```python
furnishingstatus = pd.get_dummies(df['furnishingstatus'], drop_first=True, dtype=int )
df.drop(['furnishingstatus'], inplace=True, axis=1)
df = pd.concat([df,furnishingstatus], axis=1)
```

```python
df.head()
```

```python
sns.heatmap(df.corr(), annot=True, cmap='YlGnBu')
```

```python
import statsmodels
import statsmodels.api as sm
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
df_train, df_test = train_test_split(df, test_size=0.7, random_state=100)
```

```python
scaler = MinMaxScaler()
num_cols=['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']
df_train[num_cols] = scaler.fit_transform(df_train[num_cols])
```

```python
df_train.describe()
```

```python
df_test[num_cols] = scaler.transform(df_test[num_cols])
```

```python
df_test.describe()
```

```python
# approach 1 adding 1 by 1
```

```python
y_train = df_train.pop('price')
X_train = df_train
```

```python
y_test = df_test.pop('price')
X_test = df_test
```

```python
X_train_sm = sm.add_constant(X_train['area'])
lm = sm.OLS(y_train, X_train_sm)
lr_model = lm.fit()
lr_model.summary()
```

```python
X_train_sm = sm.add_constant(X_train[['area', 'bathrooms']])
lm = sm.OLS(y_train, X_train_sm)
lr_model = lm.fit()
lr_model.summary()
```

```python
X_train_sm = sm.add_constant(X_train[['area', 'bathrooms', 'airconditioning']])
lm = sm.OLS(y_train, X_train_sm)
lr_model = lm.fit()
lr_model.summary()
```

```python
#approach 2 adding all the variables and using VIF to avoid multicolnerity
X = X_train
X_train_sm = sm.add_constant(X)
lm = sm.OLS(y_train, X_train_sm)
lr_model = lm.fit()
lr_model.summary()
```

```python
from statsmodels.stats.outliers_influence  import variance_inflation_factor
```

```python
# Calculate the VIFs again for the new model
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
#X = X_train.drop(['bedrooms'], axis=1)
#X = X_train.drop(['mainroad', 'bedrooms'], axis=1)
#X = X_train.drop(['mainroad', 'bedrooms', 'guestroom'], axis=1)
#X = X_train.drop(['mainroad', 'bedrooms', 'guestroom', 'semi-furnished'], axis=1)
X = X_train.drop(['mainroad', 'bedrooms', 'guestroom', 'semi-furnished','basement'], axis=1)
X_train_sm = sm.add_constant(X)
lm = sm.OLS(y_train, X_train_sm)
lr_model = lm.fit()
lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
#analysis
```

```python
y_train_pred = lr_model.predict(X_train_sm)
```

```python
res = y_train-y_train_pred
```

```python
sns.distplot(res)
```
