```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart_v2.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("heart disease", axis=1)
y = df["heart disease"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, random_state=42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(max_depth=3)
dt.fit(X_train, y_train)
```

```python
from IPython.display import Image
from six import StringIO
from sklearn.tree import export_graphviz
import pydotplus
#import graphviz
```

```python
dot_data = StringIO()
export_graphviz(dt, out_file = dot_data, filled = True, rounded=True, feature_names = X.columns, class_names = ['No_Dis', 'Dis'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
y_train_pred = dt.predict(X_train)
print(accuracy_score(y_train, y_train_pred))
print(confusion_matrix(y_train, y_train_pred))
```

```python
def getGraph(dt):
    dot_file = StringIO()
    export_graphviz(dt, out_file = dot_file, filled = True, rounded=True, feature_names = X.columns, class_names = ["NoDis", "Dis"])
    graph = pydotplus.graph_from_dot_data(dot_file.getvalue())
    return graph
```

```python
def getMetrics(dt_classifier):
    print("Train Accuracy :", accuracy_score(y_train, dt_classifier.predict(X_train)))
    print("Train Confusion Matrix:")
    print(confusion_matrix(y_train, dt_classifier.predict(X_train)))
    print("-"*50)
    print("Test Accuracy :", accuracy_score(y_test, dt_classifier.predict(X_test)))
    print("Test Confusion Matrix:")
    print(confusion_matrix(y_test, dt_classifier.predict(X_test)))
```

```python
Image(getGraph(dt).create_png())
```

```python
getMetrics(dt)
```

```python
#HyperParamSearch
from sklearn.model_selection import GridSearchCV
params = {
    'max_depth': [2, 3, 5, 10, 20],
    'min_samples_leaf': [5, 10, 20, 50, 100],
    'criterion': ["gini", "entropy"]
}
```

```python
dt = DecisionTreeClassifier(random_state=42)
```

```python
grid_search = GridSearchCV(estimator = dt, param_grid = params, cv=4, n_jobs=-1, verbose=1, scoring = "accuracy")
```

```python
grid_search.fit(X_train, y_train)
```

```python
score_df = pd.DataFrame(grid_search.cv_results_)
score_df.head()
```

```python
grid_search.best_estimator_
```
