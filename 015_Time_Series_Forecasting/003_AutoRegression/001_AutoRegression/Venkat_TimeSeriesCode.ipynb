## Missing Value Treatment

### Replace with average

```python
df["PassengersMissingAverage"] = df["Passengers"].fillna(df["Passengers"].mean())
```

```python
plt.figure(figsize=(12,3))
plt.plot(df["PassengersMissingAverage"])
plt.show()
```

### Replace with linear inpolation

```python
df["PassengersMissingLinear"] = df["Passengers"].interpolate(method='linear')
```

```python
plt.figure(figsize=(12,3))
plt.plot(df["PassengersMissingLinear"])
plt.show()
```

```python
# we choose linear method to handle the missing values
```

```python
df["Passengers"] = df["PassengersMissingLinear"]
df.drop(columns=['PassengersMissingLinear','PassengersMissingAverage'],inplace=True)
```

## Forecase

```python
train_len = 120
test_len = 24
train = df[:train_len]
test = df[train_len:]
```

```python
train.shape, test.shape
```

```python
## Naive
## Simple Average
## Moving Average method
## Exponential Smoothing - Simple Exponential
## Holts Exponential Smoothing
```

```python
test["Naive"] = train["Passengers"][train_len-1]
test["Average"] = train["Passengers"].mean()
ma_window = 12
test["MovingAverage"] = df['Passengers'].rolling(ma_window).mean()[train_len:]
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
#SimpleExpSmoothing
model = SimpleExpSmoothing(train['Passengers'])
model_fit = model.fit(smoothing_level=0.2,optimized=False)
model_fit.params
test['SimpleExpSmoothing'] = model_fit.forecast(test_len)
```

```python
#ExponentialSmoothing
model = ExponentialSmoothing(np.asarray(train['Passengers']) ,seasonal_periods=12 ,trend='additive', seasonal=None)
model_fit = model.fit(smoothing_level=0.2, smoothing_slope=0.01, optimized=False)
print(model_fit.params)
test['ExponentialSmoothing'] = model_fit.forecast(test_len)
```

```python
#ExponentialSmoothingOpt
model = ExponentialSmoothing(np.asarray(train['Passengers']) ,seasonal_periods=12 ,trend='add', seasonal='add')
model_fit = model.fit(optimized=True)
print(model_fit.params)
test['ExponentialSmoothingOpt'] = model_fit.forecast(test_len)
```

```python
plt.figure(figsize=(12,3))
plt.plot(train["Passengers"])
plt.plot(test["Passengers"])
#plt.plot(test["Naive"])
#plt.plot(test["Average"])
#plt.plot(test["MovingAverage"])
#plt.plot(test["SimpleExpSmoothing"])
plt.plot(test["ExponentialSmoothing"])
plt.plot(test["ExponentialSmoothingOpt"])
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
actual=test["Passengers"]
def printMet(pred):
    print("MSE", np.sqrt(mean_squared_error(actual, pred)))
    print("MAPE", np.mean(abs(actual - pred)/actual*100))
```

```python
printMet(test["Naive"])
printMet(test["Average"])
printMet(test["MovingAverage"])
printMet(test["SimpleExpSmoothing"])
printMet(test["ExponentialSmoothing"])
printMet(test["ExponentialSmoothingOpt"])
```

## AutoCorrelation

```python
from statsmodels.tsa.stattools import kpss
```

```python
kpss_model = kpss(df["Passengers"])
print("KPSS statistics ", kpss_model[0])
print("Crtitcal value @ 0.05", kpss_model[3]['5%'])
print("P-Value ", kpss_model[1])
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adfuller_model = adfuller(df["Passengers"])
print("ADF statistics ", adfuller_model[0])
print("Crtitcal value @ 0.05", adfuller_model[4]['5%'])
print("P-Value ", adfuller_model[1])
```

## Differencing

```python
from scipy.stats import boxcox
```

```python
df["PassengersBoxcox"] = boxcox(df["Passengers"], lmbda = 0)
```

```python
plt.figure(figsize=(12,4))
plt.plot(df["Passengers"])
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.plot(df["PassengersBoxcox"])
plt.show()
```

```python
df["PassengersBoxcoxDiff"] = pd.Series(df["PassengersBoxcox"] - df["PassengersBoxcox"].shift(), index =df.index )
```

```python
plt.figure(figsize=(12,4))
plt.plot(df["PassengersBoxcoxDiff"])
plt.show()
```

### Test the differentiated data

```python
kpss_model = kpss(df["PassengersBoxcoxDiff"].dropna())
print("KPSS statistics ", kpss_model[0])
print("Crtitcal value @ 0.05", kpss_model[3]['5%'])
print("P-Value ", kpss_model[1])
```

```python
adfuller_model = adfuller(df["PassengersBoxcoxDiff"].dropna())
print("ADF statistics ", adfuller_model[0])
print("Crtitcal value @ 0.05", adfuller_model[4]['5%'])
print("P-Value ", adfuller_model[1])
```

```python
# NOTE: For KPSS method the p-value is greater than 0.05 so we fail reject null hypothesis
# Null Hypothesis for KPSS: It is stationary
# For adfuller method the p-value is less then 0.05 so we can reject null hypothesis
# Null Hypothesis for adfuller: It is not stationary
```

# Auto Regression

```python
df["PassengersBoxcoxDiff"].shape, df["PassengersBoxcoxDiff"].dropna().shape
```

```python
AR_DF = df["PassengersBoxcoxDiff"].dropna()
AR_DF_train = AR_DF[:train_len-1]
AR_DF_train = AR_DF[train_len-1:]
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model = ARIMA(AR_DF_train, order=(1, 0, 0)) 
model_fit = model.fit()
print(model_fit.params)
```

```python
y_hat_ar = AR_DF.copy()
y_hat_ar['ar_forecast_boxcox_diff'] = model_fit.predict(AR_DF.index.min(), AR_DF.index.max())
y_hat_ar['ar_forecast_boxcox'] = y_hat_ar['ar_forecast_boxcox_diff'].cumsum()
y_hat_ar['ar_forecast_boxcox'] = y_hat_ar['ar_forecast_boxcox'].add(df["PassengersBoxcox"][0])
y_hat_ar['ar_forecast'] = np.exp(y_hat_ar['ar_forecast_boxcox'])
```

```python
plt.figure(figsize=(12,4))
plt.plot(train['Passengers'], label='Train')
plt.plot(test['Passengers'], label='Test')
plt.plot(y_hat_ar['ar_forecast'][test.index.min():], label='Auto regression forecast')
plt.legend(loc='best')
plt.title('Auto Regression Method')
plt.show()
```

```python
model = ARIMA(AR_DF_train, order=(0, 0, 1))
model_fit = model.fit()
print(model_fit.params)
```
