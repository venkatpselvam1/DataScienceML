```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Cricket.csv', sep=",", encoding="ISO-8859-1", header=0)
```

# Scalling

```python
from sklearn.preprocessing import StandardScaler
```

```python
df[['Mat','Inns','NO','Runs','Ave','BF','SR','100','50','0']]
```

```python
df.drop(['Player'], axis=1)
```

```python
df['HS'] = df['HS'].apply(lambda x: int(x.split('*')[0]))
```

```python
scaler = StandardScaler()
df_fmt = scaler.fit_transform(df[['SR', 'Ave']])
```

```python
df_fmt = pd.DataFrame(df_fmt)
```

```python
df_fmt
```

# Modelling

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4, max_iter = 50, random_state = 100)
kmeans.fit(df_fmt)
```

```python
kmeans.labels_
```

```python
df['label'] = kmeans.labels_
```

```python
df.to_csv('venkat.csv')
```

```python
df[df['label'] == 2]
```

# Modelling using Hierycy

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
import matplotlib.pyplot as plt
```

```python
mergings = linkage(df_fmt, method='complete', metric = 'euclidean')
dendrogram(mergings)
plt.show()
```

```python
df['hier_label'] = cut_tree(mergings, n_clusters = 4).reshape(-1, )
```

```python
df.head()
```

```python
df.to_csv('venkat.csv')
```
