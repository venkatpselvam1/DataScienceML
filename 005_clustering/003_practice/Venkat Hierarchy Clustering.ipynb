```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Main.csv', sep=",", encoding="ISO-8859-1", header=0)
```

```python
df.head()
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
from sklearn.preprocessing import StandardScaler 
```

```python
df.drop(['ï»¿Area Name'], axis=1)
```

```python
#num_cols = ['amount','frequency','recency']
scaler = StandardScaler()
std_df = scaler.fit_transform(df.drop(['ï»¿Area Name'], axis=1))
rfm_df = pd.DataFrame(std_df)
```

```python
import matplotlib.pyplot as plt
```

```python
mergings = linkage(rfm_df, method='single', metric = 'euclidean')
dendrogram(mergings)
plt.show()
```

```python
mergings = linkage(rfm_df, method='complete', metric = 'euclidean')
dendrogram(mergings)
plt.show()
```

```python
mergings = linkage(rfm_df, method='average', metric = 'euclidean')
dendrogram(mergings)
plt.show()
```
